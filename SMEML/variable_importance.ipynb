```python
import pickle
```

```python
model = pickle.load(open('model.pkl', 'rb'))
```

```python
from matplotlib import pyplot as plt
# print feature importance
features = []
for estimator in model.estimators_:
    # print the feature titles
    if len(features) == 0:
        for i, feature in enumerate(estimator.feature_importances_):
            features.append(0)

    for i, feature in enumerate(estimator.feature_importances_):
        features[i] += feature

# print feature importance as a bar chart

plt.figure(figsize=(20, 10))
plt.bar(range(len(features)), features)
# percentage of the total importance
total = sum(features)

for i, feature in enumerate(features):
    plt.text(i, feature, f'{feature/total:.4f}', ha='center', va='bottom')
plt.show()
```
